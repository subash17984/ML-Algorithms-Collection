```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform
from sklearn.svm import SVC
from sklearn.datasets import load_iris
```

```python
# Load dataset
iris = load_iris()
X, y = iris.data, iris.target
```

```python
# Define hyperparameter distribution
svc = SVC()
param_dist={
    'C':uniform(0.1,10),
    'kernel':['linear','rbf'],
    'gamma':['scale','auto']
}
```

```python
#Randomized Search with 10 iterations and 5 fold CV
random_search=RandomizedSearchCV(svc,param_distributions=param_dist, n_iter=10,cv=5,random_state=45)
random_search.fit(X,y)
```

```python
print("Best parameters:", random_search.best_params_)
print("Best cross-val score:", random_search.best_score_)
```
